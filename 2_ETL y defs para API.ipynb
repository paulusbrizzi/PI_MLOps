```python
join1 = join[join["developer"] == "Valve"]
# agrupamos por año, contamos los items y calculamos el porcentaje free
agrupado = join1.groupby('year_release').agg({
    'item_id': 'count',
    'price': lambda x: (x == 0.00).mean() * 100
    }).reset_index()
# renombramos las columnas
agrupado.columns = ['Año', 'Cantidad de Items', 'Contenido Free']
# volvemos a ordenar de manera descendente por año
agrupado = agrupado.sort_values(by='Año', ascending=False)
# formateamos el resultado
diccionario_resultado = agrupado.to_dict(orient='records')

diccionario_resultado
```

```python
reviews = reviews[["item_id","recommend", "analisis_sentimiento", "year_posted"]]
reviews

# trabajamos con nuestras columnas de interés y formamos una nueva tabla
games = games[["item_id", "developer"]]
reviews = reviews[["item_id","recommend", "analisis_sentimiento", "year_posted"]]
merge = pd.merge(games, reviews, on="item_id", how="inner")
merge = merge[["item_id", "developer", "recommend", "year_posted"]][merge["analisis_sentimiento"] == 2]
merge = merge[["year_posted", "developer"]][merge["recommend"] == 1]
merge

# Obtener los 3 géneros más comunes para cada año
resultados = []
for año, grupo in merge.groupby('year_posted'):
    top3_devs = grupo['developer'].value_counts().nlargest(3).index.tolist()
    for dev in top3_devs:
        resultados.append({str(año): dev})

resultados

# Convertir la lista de diccionarios a un DataFrame
df = pd.DataFrame([(k, v) for d in resultados for k, v in d.items()], columns=['Año', 'Desarrollador'])

# Imprimir el DataFrame resultante
df

df.to_csv("apalala_api/best_developer_year.csv")
```

```python
# trabajamos con nuestras columnas de interés y formamos una nueva tabla
games = games[["item_id", "developer"]]
reviews = reviews[["item_id", "analisis_sentimiento"]]
merge = pd.merge(games, reviews, on="item_id", how="inner")
merge = merge[["developer", "analisis_sentimiento"]]
merge

# Crear nuevas columnas con el conteo de valores 0 y 2
merge['Negative'] = (merge['analisis_sentimiento'] == 0).astype(int)
merge['Positive'] = (merge['analisis_sentimiento'] == 2).astype(int)

# Agrupar por valores de la columna 1 y sumar los conteos
agrupado = merge.groupby('developer')[['Negative', 'Positive']].sum().reset_index()

# Imprimir el resultado
agrupado["developer"] = agrupado["developer"].str.lower()

agrupado

```

```python
agrupado.to_csv("data_api/developer_reviews_analysis.csv")
```
