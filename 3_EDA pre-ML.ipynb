```python
sns.boxplot(x='playtime_forever',y="year_release",data=items)
plt.show()
```

```python
# acá podemos observar que existe una marcada curva de crecimiento de juegos lanzados a partir del año 1990.

# se puede ver también que el gráfico mantiene horas de juego muy variables, se puede interpretar
# que existen muchos perfiles de usuarios que suelen disfrutar de juegos con distintos tiempos acumulados,
# y que estos juegos mayoritariamente son los lanzados a partir del año 2003 aproximadamente.

# se podría hipotetizar que no existe gran cantidad de horas acumuladas para juegos lanzados alrededor
# del año 1997 porque podría tratarse de juegos con poco desarrollo de historia a comparación
# de los más contemporáneos.
```
